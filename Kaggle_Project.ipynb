```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
url="https://stats.espncricinfo.com/ci/content/records/284269.html"
```

```python
temp=pd.read_html(url)
```

```python
temp
```

```python
cr=temp[0]
```

```python
cr.head()
```

```python
cr.info()
```

```python
mostducks=cr[['Player', '0']].sort_values(by="0",ascending=False)

plt.figure(figsize=(8,8))
graph=sns.barplot(y='Player',palette='rocket',x='0',data=mostducks.head(25))
graph.set_title('Most Ducks By a Batsman')
plt.show()
```
